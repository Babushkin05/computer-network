```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```python
df = pd.read_csv('client/case1.csv')

plt.plot(df['bytes'], df['avg_ms'], marker='o', linewidth=2)

plt.xlabel('  (bytes)', fontsize=12)
plt.ylabel('  (avg_ms)', fontsize=12)
plt.title('N=8 M=5000 Q=25', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend([''], loc='best')

plt.tight_layout()
plt.show()
```
